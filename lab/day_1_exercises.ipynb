```python
import numpy as np
```

```python
def mat_mult(A, B):
    ## multiply matrices: A * B = C 
    ## return C

    ## the number of columns in the first matrix must equal the number of rows in the second matrix.
    if (A.shape[1] != B.shape[0]):
        print("The number of columns in the first matrix must equal the number of rows in the second matrix.")
        return None
    

    m, n = A.shape[0], B.shape[1]
    C = np.empty((m, n))

    for i in range(0, m):
        for j in range(0, n):
            s = 0
            for k in range(0, A.shape[1]):
                s += (A[i, k] * B[k, j])
            C[i, j] = s
    
    return C
```

```python
def test_mat_mult(mat_mult):
    # Basic 2x2
    A = np.array([[1, 2], [3, 4]])
    B = np.array([[5, 6], [7, 8]])
    assert np.allclose(mat_mult(A, B), np.matmul(A, B)), "Failed on 2x2"

    # Non-square: (2x3) @ (3x2)
    A = np.array([[1, 2, 3], [4, 5, 6]])
    B = np.array([[7, 8], [9, 10], [11, 12]])
    assert np.allclose(mat_mult(A, B), np.matmul(A, B)), "Failed on (2x3) @ (3x2)"

    # Identity matrix
    A = np.array([[3, -1], [0, 7]])
    I = np.eye(2)
    assert np.allclose(mat_mult(A, I), A), "Failed on identity"

    # Column vector: (3x3) @ (3x1)
    A = np.array([[1, 0, 2], [0, 1, 0], [3, 0, 1]])
    v = np.array([[1], [2], [3]])
    assert np.allclose(mat_mult(A, v), np.matmul(A, v)), "Failed on matrix @ column vector"

    # Random large matrices
    A = np.random.randn(10, 20)
    B = np.random.randn(20, 15)
    assert np.allclose(mat_mult(A, B), np.matmul(A, B)), "Failed on random (10x20) @ (20x15)"

    print("All tests passed.")

test_mat_mult(mat_mult)
```
